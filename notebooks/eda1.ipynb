```python
# 📚 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# 📂 2. Load the processed stock data
df = pd.read_csv('../data/processed/AAWW_processed.csv')  # <-- adjusted path

# 🧹 3. Basic data check
print("Shape of data:", df.shape)
print("\nData types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())

# 🧠 4. Statistical summary
print("\nStatistical summary:\n", df.describe())

# 🛑 5. Check for missing values
print("\nMissing values per column:\n", df.isnull().sum())

# 📈 6. Visualize Close price over time
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'])
plt.title('AAWW - Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 📊 7. Moving Averages (5d and 20d)
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['MA_5d'], label='5-Day MA')
plt.plot(df['Date'], df['MA_20d'], label='20-Day MA')
plt.title('AAWW - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 📈 8. Volatility visualization
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Volatility_5d'], label='5-Day Volatility')
plt.plot(df['Date'], df['Volatility_20d'], label='20-Day Volatility')
plt.title('AAWW - Volatility over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 📉 9. MACD and MACD Signal
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['MACD'], label='MACD')
plt.plot(df['Date'], df['MACD_Signal'], label='MACD Signal')
plt.title('AAWW - MACD Indicator')
plt.xlabel('Date')
plt.ylabel('MACD Value')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 📈 10. RSI visualization
plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['RSI_14d'], label='RSI 14d', color='purple')
plt.axhline(70, color='red', linestyle='--')  # Overbought threshold
plt.axhline(30, color='green', linestyle='--')  # Oversold threshold
plt.title('AAWW - RSI (14 days)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()


# ✅ Done with EDA!
```
